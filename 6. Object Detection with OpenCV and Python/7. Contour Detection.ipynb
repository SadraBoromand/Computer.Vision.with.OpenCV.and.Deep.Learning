```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/internal_external.png',0)
```

```python
img.shape
```

```python
plt.imshow(img,cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contours)
```

```python
len(contours)
```

```python
type(hierarchy)
```

```python
hierarchy
```

```python
external_conturs = np.zeros(img.shape)
```

```python
external_conturs.shape
```

```python
list(range(len(contours)))
```

```python
for i in range(len(contours)):
    
    # EXTERNAL
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(external_conturs,contours,i,255,-1)
```

```python
plt.imshow(external_conturs,cmap='gray')
```

```python
internal_conturs = np.zeros(img.shape)

for i in range(len(contours)):
    
    # INTERNAL
    if hierarchy[0][i][3] != -1:
        
        cv2.drawContours(internal_conturs,contours,i,255,-1)
```

```python
plt.imshow(internal_conturs,cmap='gray')
```

```python
peperony_conturs = np.zeros(img.shape)

for i in range(len(contours)):
    
    # INTERNAL
    if hierarchy[0][i][3] == 4:
        
        cv2.drawContours(peperony_conturs,contours,i,255,-1)
        
plt.imshow(peperony_conturs,cmap='gray')
```

```python
face_conturs = np.zeros(img.shape)

for i in range(len(contours)):
    
    # INTERNAL
    if hierarchy[0][i][3] == 0:
        
        cv2.drawContours(face_conturs,contours,i,255,-1)
        
plt.imshow(face_conturs,cmap='gray')
```
